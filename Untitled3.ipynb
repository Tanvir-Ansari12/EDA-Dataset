```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Titanic-Dataset.csv')

print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns='Cabin', inplace=True)
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
numerical_cols = ['Age', 'Fare']
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['Age'])
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Fare'])
plt.show()
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]
df = remove_outliers(df, 'Age')
df = remove_outliers(df, 'Fare')
```

```python
print(df.describe())
print(df.describe(include=['O']))
print("Median Age:", df['Age'].median())
print("Fare IQR:", df['Fare'].quantile(0.75) - df['Fare'].quantile(0.25))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
numeric_cols = ['Age', 'Fare']
df[numeric_cols].hist(bins=20, figsize=(8,4))
plt.suptitle('Histograms of Age and Fare')
plt.show()
plt.figure(figsize=(8,4))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, 2, i)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', diag_kind='hist')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
corr = df.select_dtypes(include='number').corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
print(df.groupby('Pclass')['Survived'].mean())
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival Rate by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Count')
plt.show()
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

```python
def reverse_embarked(row):
    if row['Embarked_S'] == 1:
        return 'S'
    elif row['Embarked_Q'] == 1:
        return 'Q'
    else:
        return 'C'
df['Embarked'] = df.apply(reverse_embarked, axis=1)
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival Rate by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Count')
plt.show()
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```
